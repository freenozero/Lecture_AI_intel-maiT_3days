 -> colab 기준 해당 mp3 오픈시에 '반갑습니다'라고 함

```python
from gtts import gTTS
import speech_recognition
import playsound
import os
import time

def speak(text):
    tts = gTTS(text=text, lang='ko')
    filename='voice.mp3'
    tts.save(filename)
    playsound.playsound(filename)
speak('반갑습니다!')
```
